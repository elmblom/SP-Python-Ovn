```python
# Script programering 3-15

x = 40

while x != -41:
    y = round((x * 9/5) + 32,1)
    print("{}C -> {}F".format(x, y))
    x -= 1
```

```python
# Script programering 3-17

PN = str(input("Please eneter your swedish social security number: "))

if len(PN) == 11:
    format = 0
elif len(PN) == 13:
    format = 1
else:
    print("SSN is wrong length")
    exit
    
if format == 0:
    if PN[6] != '-':
        print("SSN is wrong format")
        exit
    else:
        PN = PN.replace("-", "")
elif format == 1:
    if PN[8] != '+':
        print("SSN is wrong format")
        exit
    else:
        PN = PN.replace("+", "")

lastnum = (PN[ -1: ])
others = (PN[ : -1 ])

total = 0 

for i, num in enumerate(others):
    num = int(num)
    if i % 2 == 0:
        doubled = num * 2
        total += doubled if doubled < 10 else doubled - 9
    else:
        total += num

checksum = (10 - (total % 10)) % 10
if checksum == int(lastnum): 
    print("SSN valid!")
else:
    print("SSN not valid!")


```

```python
# Script programering 3-18

# pip install afinn

from afinn import Afinn

af = Afinn()

response = str.lower((input("How are you doing today?")))

score = af.score(response)

if score > 5:
    print("Wow, you seem to be having an amazing day! Keep shining!")
elif score > 0:
    print("That's great to hear! Keep up the positive vibes!")
elif score < -5:
    print("I'm really sorry to hear that you're feeling this way. Please take care and reach out to someone if you need help.")
elif score < 0:
    print("Oh no, I'm sorry to hear that. I hope things get better soon!")
else:
    print("I didn't quite understand that. Could you please rephrase?")
```
